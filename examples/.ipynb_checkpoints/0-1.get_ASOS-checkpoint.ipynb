```python
save_columns = [
    'true_ghi', 'true_ghi_corrected',
    'azimuth', 'zenith',
    'cs_ghi', 'cs_dni', 'cs_dhi','cloud_index',
    'stn_id','name','latitude', 'longitude','altitude'
]
```

```python
asos_ghi = pd.DataFrame()
for i, row in asos_meta.iterrows():
    stn_nm = row.stn_nm
    print(stn_nm)
        
    lat, lon, alt = row.latitude, row.longitude, row.altitude
    stn_loc = Location(lat, lon, altitude=alt, tz='Asia/Seoul')

    cur_asos = asos_data[['stn_id','stn_nm', 'true_ghi','latitude', 'longitude','altitude','cloud_index']][asos_data.stn_nm == stn_nm]
    
    # convert MJ -> Watt (just converting unit in this step)
    
    solpos = stn_loc.get_solarposition(cur_asos.index)
    cs_irrads = stn_loc.get_clearsky(cur_asos.index, solar_position=solpos)
    cs_irrads.columns = [f'cs_{c}' for c in cs_irrads.columns]
    cur_asos = cur_asos.join(solpos).join(cs_irrads)
    
    # replace irradiance of np.nan value measured at hour H with 0
    # if solar elevation for that time is less than 0
    cur_asos.loc[cur_asos.true_ghi.isnull() & (cur_asos.elevation < 0), 'true_ghi'] = 0.
    cur_asos = cur_asos.sort_index()
    
    # calculates gradient of the cumulative irradiance function
    # to get the instantaneous power output for each hour
    ghi_cumsum = cur_asos.groupby(cur_asos.index.date)['true_ghi'].cumsum()
    ghi_deriv = ghi_cumsum.groupby(ghi_cumsum.index.date).\
        apply(lambda x: pd.DataFrame(
            np.gradient(x, x.index.hour,edge_order = 1), index=x.index) if len(x)>1 else pd.DataFrame(
            [np.NaN], index=x.index))[0].\
        rename('true_ghi_corrected')
    
    cur_asos = cur_asos.join(ghi_deriv)
    cur_asos['name'] = stn_nm
    # if stn_id[0] == '0':
    #     cur_asos.index -= pd.Timedelta(hours=1)
    asos_ghi = asos_ghi.append(cur_asos[save_columns])
# asos = asos_ghi.copy()
asos_ghi['true_ghi_corrected'] *= 277.7777
```

```python
asos_ghi = asos_ghi.drop_duplicates()
asos_ghi = asos_ghi[asos_ghi.true_ghi_corrected>0]

raw_asos_ghi = asos_ghi
CS_DIR = './'
raw_asos_ghi.to_pickle(CS_DIR+'cs_ghi.pkl')
```

```python
asos_ghi = raw_asos_ghi.copy()
asos_ghi['Cloud_OD'] = (asos_ghi.true_ghi_corrected-asos_ghi.cs_dhi)/(asos_ghi.cs_dni)
asos_ghi = asos_ghi[(asos_ghi['Cloud_OD']>=0)&(asos_ghi['Cloud_OD']<=1)]
asos_ghi['cloud_index'] *= 0.1
asos_ghi['cloud_index'] = asos_ghi.cloud_index+0.05
asos_ghi['cloud_index'] = asos_ghi['cloud_index'].clip(0,1) 
```

```python
asos_ghi['cal_Cloud_OD'] = ((1-asos_ghi.cloud_index)/((1/np.cos(np.pi/180*asos_ghi.zenith)) - (asos_ghi.cloud_index))).clip(0,1)
```

```python
ghi_test_set = []
errors = np.zeros(11)
ghi_OBS = pd.DataFrame()

for z,cloud_index in enumerate(np.arange(0,1.01,0.1)):
    
    print('='*20)
    print(f'cloud index      =      {cloud_index}')

    cal_cloud = cloud_index
    cal_cloud +=0.05
    cal_cloud = cal_cloud.clip(0,1)
    switch = True
    
    if switch :  trs  =  0.025
    else      :  trs  =  100000
    
    
    
    tmp_ghi = asos_ghi[asos_ghi.zenith<90].copy()
    tmp_ghi = tmp_ghi[abs(tmp_ghi.cloud_index - cal_cloud)<trs]
    tmp_ghi.cal_Cloud_OD = tmp_ghi.cal_Cloud_OD.round(4)
    tmp_ghi = tmp_ghi.dropna()
    
    ghi_OBS = ghi_OBS.append(tmp_ghi[['name','stn_id','latitude','longitude','true_ghi_corrected']])
    
    #==========================================================================================================================================
    
    model_col = ['azimuth', 'zenith','cs_ghi','cs_dni', 'cs_dhi', 'cloud_index','latitude','longitude', 'altitude', 'cal_Cloud_OD', 'Cloud_OD']
    ghi_to, ghi_test ,__ ,__ = get_train_test_data(tmp_ghi[model_col], 
                        start_date='2020-01-01',end_date='2022-12-31', split_freq=5)
    ghi_test_set.append(ghi_test)
    
    ghi_model = setup(session_id=1234, train_size=0.8, data = ghi_to, target = 'Cloud_OD', normalize=True, transformation=True, use_gpu=False, silent =True)
    ghi_model_res = compare_models(n_select=1,sort = 'MAE', include=['lightgbm'])
    # top5 = ghi_model_res[0:5]
    # tuned_top5 = [tune_model(i) for i in top5]
    # blender_top5 = blend_models(estimator_list=top5)
    tuned_model = tune_model(ghi_model_res)
    final_model = finalize_model(tuned_model)
    
    save_model(final_model, f'/mnt/sdb1/wscho/Jobs/gk2aICT/ca_model_index_{z}')
    
    # with open(f'/mnt/sdb1/wscho/Jobs/gk2aICT/ca_model_index_{z}.pkl',"wb") as f:
    #     pickle.dump(final_model, f)
        
    prediction = predict_model(final_model, data=ghi_test)
    # prediction = predict_model(final_model, data=ghi_test)

    y = prediction.Label
    x = prediction.Cloud_OD
    errors[z] = np.mean(abs(y-x))
    
    print(errors[z])
```
